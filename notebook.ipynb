```python
import pandas as pd
```

```python
data=pd.read_csv("House Price India.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
stats=data["Price"].describe().reset_index()
stats["Price"]=round(stats["Price"],2)
```

```python
data.duplicated().sum()
```

```python
# for na values
data.dropna(inplace=True)
# for duplicate values
data.drop_duplicates(inplace=True)
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
data.groupby("condition of the house")["Price"].mean().sort_values(ascending=False).plot(kind="bar")
plt.title("condition of the house vs Price")
plt.ylabel("Mean Price")
plt.xlabel("Condition of the house")
plt.show()
```

```python
data.columns
```

```python
x= data[['number of bedrooms', 'number of bathrooms','living area','condition of the house','Number of schools nearby' ]]
y=data[['Price']]
```

```python
x
```

```python
y
```

```python
data.shape[0]*0.8
#this means 20% of the data will be used for testing and 80% for training   
```

```python
#train test split by scikit-learn
#used to split the data into training and testing sets
#bcz we want to train our model on one part of the data and test it on another part to evaluate its performance
# here test_size=0.2 means 20% of the data will be used for testing and 80% for training
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid={'criterion':["mse", "friedman_mse", "mae"],
            'splitter':["best", "random"],
            'max_depth':[None,10,20,30,40,50],
            'min_samples_split':[2,5,10],
            'min_samples_leaf':[1,2,4]
           }
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator=tree_model, param_grid=param_grid)
```

```python
#grid_tree.fit(x_train, y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds = grid_tree.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
preds = lr.predict(x_test)
```

```python
mean_absolute_error(y_test, preds)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr={"max_depth":[5,10,15],
            "n_estimators":[2,3,4,5,6,7,8,9,10]
              }
```

```python
gridrfr = GridSearchCV(rfrmodel, param_gridrfr)
```

```python
gridrfr.fit(x_train, y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
rfrpredictions = gridrfr.predict(x_test)
```

```python
mean_absolute_error(y_test, rfrpredictions)
```
